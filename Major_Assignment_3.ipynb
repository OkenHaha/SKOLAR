```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage import io
from google.colab import files
```

```python
a = files.upload()
```

```python
image = io.imread('fb profile.jpg')
```

```python
image_vector = image.reshape(-1, 3)
```

```python
pca = PCA(n_components=2)
pca.fit(image_vector)
```

```python
transformed_image = pca.transform(image_vector)
print(f"Original Shape: {image_vector.shape}")
print(f"Reduced Shape: {transformed_image.shape}")
```

```python
reconstructed_image = pca.inverse_transform(transformed_image)
reconstructed_image = reconstructed_image.astype(np.uint8).reshape(image.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(image)
ax1.title.set_text('Original Image')
ax2.imshow(reconstructed_image)
ax2.title.set_text('Reconstructed Image')
plt.show()
```

```python
mse = ((image - reconstructed_image) ** 2).mean(axis=None)
print(f"Mean squared reconstruction error: {mse}")
```
